Importing the necessary libraries

```python
import pandas as pd
import matplotlib as plt
```

Assigning the necessary dataframes

```python
df = pd.read_csv('https://raw.githubusercontent.com/KeithGalli/lego-analysis/master/datasets/lego_sets.csv')
```

```python
theme = pd.read_csv('https://raw.githubusercontent.com/KeithGalli/lego-analysis/master/datasets/parent_themes.csv')
```

```python
df.head()
```

```python
theme.head()
```

Joining the two dataframes by a single common column for ease of analysis. We use the parent theme column as it is available in both datasets

```python
merge_df = df.merge(theme, left_on='parent_theme', right_on='name')
merge_df.head()
```

```python
merge_df.drop(columns='name_y', inplace=True)
merge_df.head()
```

Q1 - What percentage of all licensed sets ever released were Star Wars Themed?

```python
license = merge_df[merge_df['is_licensed']]
```

Filtering out only the licensed data sets

```python
license.head()
```

```python
license['set_num'].isnull().values.sum()
```

```python
license = license.dropna(subset='set_num')
```

```python
license['set_num'].isnull().values.sum()
```

```python
license.head()
```

```python
star_wars_df = license[license['parent_theme']=='Star Wars']
```

```python
Tot_L = int(license['parent_theme'].count())
Tot_L
```

```python
Tot_S = int(star_wars_df['parent_theme'].count())
Tot_S
```

```python
the_force = int(Tot_S / Tot_L * 100)
the_force
```
